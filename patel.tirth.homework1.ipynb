* Start the program by selecting the cell and press ctrl+enter.

CODE INSTRUCTIONS-

    Enter the value of priciple: 2350000.00

    Enter the value of the year: 3

    The value of simple interest is:  680325.0

```python
def simpleinterest(principle,interest,year):                                        #defined a function to calculate the simpleinterest
       return  ((principle*interest*year)/100)

try:
    principle = float(input("Enter the value of priciple: "))                        #checked if the values entered are as per requirement
    interest = float(input("Enter the value of interest rate: "))
    year = int(input("Enter the value of the year: "))
    
    
    if (principle>=0.0 and year>0 and interest>=0.0 and interest<=100.00):
        
        si = simpleinterest(principle,interest,year)                                #function called including the 3 arguments
        
        print("\nThe value of simple interest is: ",si)

    else:
        print("\nThe entered values for either principle, interest or year is not valid")
        
except:
    print("\nThe entered values for either principle, interest or year is not valid")   #if the entered values arent according to the required datatype
```
