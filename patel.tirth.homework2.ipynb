## Problem 1

### ALOGRITHM: 
1. Take input from the user, store it in a variable
2. Computer randomly selects a choice between rock, paper, scissors and store it in a variable
3. Compare the choices made by user and computer by comparing varibales
4. Choose who wins
5. Print user's input, computer's input and who won

### CODE INSTRUCTIONS:

* Start the program by selecting the cell and press ctrl+enter.
* The program will run upto one iteration after which it will stop. The program has to be executed again to get results for different set of values.

### EXPECTED RESULT:

    Computer chooses scissors
    You, the user, win!

    Enter your choice: paper
    Computer chooses rock
    I, the computer, win!

    Enter your choice: scissors
    Computer chooses scissors
    User and computer tie!

```python

#first computer makes a choise, after that user inputs their choise. After that I've compared them and showed the result.

from random import randint

chosen = randint(1,3)      #Computer's random choise it made

player = input('Please make your choise: rock, paper or scissors ? ') #Player makes their choise.

if(chosen == 1):
    computer = 'rock'
    print('Computer chooses rock')

elif(chosen == 2):
    computer = 'paper'
    print('Computer chooses paper')

else:
    computer = 'scissors'
    print('Computer chooses scissors')
    
#-------------------------------------------------------

if(player == computer):
    print('User and computer tie!')

elif(player == 'rock' and computer == 'scissors'):
    print('You, the user, win!')

elif(player == 'rock' and computer == 'paper'):
    print('I, the computer, win!')

elif(player == 'paper' and computer == 'rock'):
    print('You, the user, win!')

elif(player == 'paper' and computer == 'scissors'):
    print('I, the computer, win!')

elif(player == 'scissors' and computer == 'paper'):
    print('You, the user, win!')

elif(player == "scissors" and computer == 'rock'):
    print('I, the computer, win!')

else:
    print('Something went wrong.')

```

### CODE INSTRUCTIONS:

### EXPECTED RESULT:

    Enter a string: mom
    “mom” is a palindrome.
    
    Enter a string: apple
    “apple” is not a palindrome.
    
    Enter a string: A nut for a jar of tuna
    “A nut for a jar of tuna” is a palindrome.
    
    Enter a string: A nut for a jar of salmon
    “A nut for a jar of salmon” is not a palindrome.


```python
def sentencePalindrome(s): 
    l, h = 0, len(s) - 1
    
    s = s.lower()                                                  # Lowercase string 
     
    while (l <= h):                                                # Compares character until they are equal
    
        
        if (not(s[l] >= 'a' and s[l] <= 'z')): 
            l += 1
    
        elif (not(s[h] >= 'a' and s[h] <= 'z')): 
            h -= 1
     
        elif (s[l] == s[h]):                                       # If characters are equal
            l += 1
            h -= 1
        
                                                                    # If characters are not equal then 
                                                                    # sentence is not palindrome 
        else: 
            return False
    
    return True                                                     # Returns true if sentence is palindrome 


s=input("Enter a string:")

a=sentencePalindrome(s)

if (a==True): 
    print (s,"is a palindrome.")
else: 
    print (s,"is not a palindrome.")
```

## Program 5

### ALGORITHM:

1. Take file name as input from the user and then store in file_name
2. Open file in read mode and store in open_file.
3. Initialize line_count,word_count and character_count
4. Run a loop to count number of lines in open_file and store it in line_count
5. Split the lines into words using split() function and run a loop to count the number of words and store it in word_count
6. Using len(word) count the number of characters.
7. Using avg = character_count/word_count, we calculate the average length of a word
8. Print line_count,word_count,character_count and avg

### CODE INSTRUCTIONS:

1. Start the program by selecting the cell
2. Program will stop after one iteration. To obtain different set of values, run thr program again.

### EXPECTED RESULT:

    What is the filename: RobertFrost.txt
    Number of lines: 4
    Number of words: 27
    Number of characters: 103
    Average length of word: 3.814814814814815

```python
try:
    file_name = input("What is the filename: ")       # enter filename
    open_file = open(file_name,"r")                   # read filename in read mode.
    line_count = 0                                    # line count
    word_count = 0                                    # word count
    character_count = 0                               # character count
    
    for line in open_file:                            # check for line count 
        line_count = line_count + 1              
                      
        for word in line.split():
            word_count = word_count + 1
            character_count = character_count + len(word) #sum(len(word) for word in open_file) 
                      
    avg = character_count/word_count                   # Calculate Average length of word = length of word/ Number of words        
    print("Number of lines:", line_count)
    print("Number of words:", word_count)
    print("Number of characters:", character_count)
    print("Average length of word:", avg)
                      
except:
    print("Enter valid file name.")
```
